Label Enconding

label enconding and categorical encoding are two techniques used to ecode categorical data as numerical data

# label encoding involves assining a unique numerical label to each category in the variable.

# labels are assigned in alphabetical order or based on the frequency of the categories 

# categorical variables are:
1. Red : 1
2. Blue : 2
3. Green : 3  
# ass unique values

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'color': ['red','green', 'blue', 'green', 'red']
})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['green']])
```


Label Encoding (problem for nominal data)

It assigns numbers:

Red → 3
Blue → 2
Green → 1

Now ML model may think:

Red > Blue > Green ❌ (fake order created)

Even though colors have no order

# thats why we use now ordinal encoding

Ordinal Encoding 

# tech assign numerical value based on its position in the order 
# categorical data have ranking 

# categorical variable with education level

1. High school : 1
2. College : 2 
3. Graduate : 3
4. post Graduate : 4


# it have ranks are impt


```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size' : ['small', 'medium', 'large', 'medium', 'small', 'large']
})
df
```

```python
## assigns ranking of each : create instance of ordinalencoder 
# and then fit_trnasform
encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
encoder.fit_transform(df[['size']])  # so here gives rank small 0 , medium 1, large 2
```
