Target Guided Ordinal Encoding 

# it is a technique used to encode categorical variables based on their relationship with target variable 

# useful when we have a categorical variable with large number of unique categories

# we replace each categorical varizble with a numerical value based on the mean or median of the target variable for the category. 

# cretes monotonic relationshop between the categorical variables and target varibles 

```python
import pandas as pd

df = pd.DataFrame({
    'city': ['new york', 'london', 'paris', 'tokyo', 'new york', 'paris'],
    'price' : [200, 150, 300, 250, 180, 320]
})
```

```python
df
```

##  based on group of city replace price with there mean or median 

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)    ## replaced with maped values
```

```python
df
```

```python
df[['city_encoded', 'price']]
```

## Based on dataset tips  : covert the time based on total bill

```python
import seaborn as sns
df = sns.load_dataset('tips')
df

```

```python
import pandas as pd
mean_totalbill = df.groupby('time')['total_bill'].mean()

mean_totalbill
```

```python
df['time_encoded'] = df['time'].map(mean_totalbill)
df
```

```python
df[['time', 'time_encoded']]
df.head()
```
