Nominal / one hot encoding

represent categorical data into numerical data 

disadvantages  : dont use when categories are more can cause overfitting
create sparse matrix can be confusing  : overfitting

1. red [1,0,0]
2. blue [0,1,0]
3. green [0,0,1]


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color': ['red', 'orange', 'purple', 'green', 'blue']
})
df.head()
```

```python
## Crete an instance of Onehotencoder

encoder = OneHotEncoder()
```

```python
## perform fit and transform

encoded = encoder.fit_transform(df[['color']]).toarray()  ## doble bracket for color gives in rows and col value 
```

```python
import pandas as pd
encoded_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())  # automatic col names created 
```

```python
encoded_df
```

```python
encoder.transform([['blue']]).toarray()   #check for single 
```

```python
import seaborn as sns

df = sns.load_dataset('tips')
df                 ### convert this categorical variables sex, smoker, day, time to onehotencoder
```

```python
df[['sex', 'smoker', 'day', 'time']]
```

```python

from sklearn.preprocessing import OneHotEncoder
```

```python
# create Instance

encoder = OneHotEncoder()
```

```python
# perform fit_transform we can convert categories into variables

encoded = encoder.fit_transform(df[['sex']]).toarray()
```

```python
encoded
```

```python
## Create data frame to check the categories are encoded or not using pandas
```

```python
import pandas as pd

encoded_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```
